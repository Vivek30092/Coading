```python
#place the target data price in a separate dataframe y_data
y_data=df['price']
```

```python
# drop price data in dataframe x_data
x_data=df.drop('price',axis=1)
```

```python
#now randomly split data into trining and testing data using the function train_test_split
from sklearn.model_selection import train_test_split


x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.10, random_state=1)


print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
# importing linearRegression

from sklearn.linear_model import LinearRegression
lre=LinearRegression()

# fiting the model using feature 'horsepower'
lre.fit(x_train[['horsepower']],y_train)
```

```python
# calculating R^2 on the test data
lre.score(x_test[['horsepower']],y_test)
```

```python
lre.score(x_train[['horsepower']],y_train)
```
