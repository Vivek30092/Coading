```python
df['normalized-losses'].replace("?",np.nan,inplace=True)
print(df['normalized-losses'].astype(float))
```

```python
avg_nor_loss=df['normalized-losses'].astype('float').mean(axis=0)
df['normalized-losses']=df['normalized-losses'].replace(np.nan,avg_nor_loss)
print(df['normalized-losses'])
```

```python
df['bore'].replace("?",np.nan,inplace=True)
avg_bor=df['bore'].astype("float").mean(axis=0)
df['bore']=df['bore'].fillna(avg_bor,inplace=True)
df['bore'].replace(np.nan,avg_bor)
```

```python
df.replace('?',np.nan,inplace=True)
mean_value=df['stroke'].astype("float").mean(axis=0)
df['stroke'].replace(np.nan,mean_value)
```

```python
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
df['horsepower'].replace(np.nan,avg_horsepower,inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
df['peak-rpm']=df['peak-rpm'].replace(np.nan,avg_peakrpm,inplace=True)
```

```python
df['num-of-doors'].value_counts().idxmax()
df['num-of-doors'].replace(np.nan,'four')
```

```python
df.dropna(subset=['price'],axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
```

**binning**

```python
df['horsepower']=df['horsepower'].astype(int,copy=True)
```

creating histogram of horsepower


```python
import matplotlib as plt 
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
import numpy as np
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
group_names=['low','medium','high']
df['horsepower-binned']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(5)
```

```python
df['horsepower-binned'].value_counts()
```

plot for the distribution of each bin.


```python
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names,df['horsepower-binned'].value_counts())

#set x/y labels and plot title
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```
